```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
```

```python
df = pd.read_csv('results/results_all5.csv')
df.head()
```

```python
replacable_str = ("<class 'NiaPy.algorithms.", "'>", "basic.gwo.", "modified.jde.",'basic.ga.', 'basic.es.', 'basic.pso.')

for s in replacable_str:
    df['Algorithm'] = df['Algorithm'].str.replace(s,'')

df.head()
```

```python
df['Algorithm'].unique()
```

```python
df['Meta'] = None
df.loc[df['Algorithm'].str.contains("DynFS"), 'Meta'] = ' DynFS'
df.loc[df['Algorithm'].str.contains("EvoFS"), 'Meta'] = ' EvoFS'

df['Nia'] = None
df.loc[df['Algorithm'].str.contains("GreyWolfOptimizer"), 'Nia'] = ' GWO'
df.loc[df['Algorithm'].str.contains("SelfAdaptiveDifferentialEvolution"), 'Nia'] = ' saDE'
df.loc[df['Algorithm'].str.contains("GeneticAlgorithm"), 'Nia'] = ' GA'
df.loc[df['Algorithm'].str.contains("EvolutionStrategyMpL"), 'Nia'] = ' ES'
df.loc[df['Algorithm'].str.contains("ParticleSwarmAlgorithm"), 'Nia'] = ' PSO'
df.head()
```

```python
df.to_csv('data.csv')
```
